```python
url = 'https://raw.githubusercontent.com/nishant-harinkhede12/Advanced-M-C-projects/main/AAPL.csv'
import numpy as np
import pandas as pd
import plotly.graph_objects as go
data = pd.read_csv(url)
print(data.head())
```

```python
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], high=data["High"],
                                        low=data["Low"], close=data["Close"])])
figure.update_layout(title = "Apple Stock Price Analysis", xaxis_rangeslider_visible=False)
figure.show()
```

```python
print(data.corr())
```

```python
from autots import AutoTS
model = AutoTS(forecast_length=15, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```

```python
print(forecast)
```
